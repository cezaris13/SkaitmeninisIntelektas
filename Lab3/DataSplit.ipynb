```python
# import from scikit learn iris dataset, split it to train and test data, and print the data
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import numpy as np
iris = load_iris()
X = iris.data
y = iris.target
# np.set_printoptions(precision=3)
# print("X: ", X)
# get ys as names
# y_names = iris.target_names[y]
# print("y_names: ", y_names)
# split the data to train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y)


printData(X_train, y_train)
print("=====================================")
printData(X_test, y_test)

print ("y_train: ", np.bincount(y_train))
print ("y_test: ", np.bincount(y_test))

def printData(X, y):
    for i in range(len(X)):
        print(str(X[i]).replace('[', '').replace(']', '').replace(" ",",").replace(".,",".0"), end=",")
        print (iris.target_names[y[i]])
```
